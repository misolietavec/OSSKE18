### Chceme graf - koľko ľudí sa v aký deň v roku narodilo? Osobitne muži, ženy a aj spolu.

```python
NarMuz = DniNar[DniNar['Pohlavie'] == 'muž'][['Deň','Počet']]
NarZen = DniNar[DniNar['Pohlavie'] == 'žena'][['Deň','Počet']]

NarZen = NarZen[NarZen["Deň"] != "29. február"]
NarMuz = NarMuz[NarMuz["Deň"] != "29. február"]
```

```python
fig = go.Figure(layout={'width':1000, 'height':600, "title": "Dáta narodenia Prešovčanov"})
fig.add_scatter(x=NarMuz['Deň'],
                y=NarMuz['Počet'],
                mode='markers',
                name = "Muži",
                marker={'size': 10,
                        'color': 'red',
                        'opacity': 0.6,
                       });
fig.add_scatter(x=NarZen['Deň'],
                y=NarZen['Počet'],
                mode='markers',
                name="Ženy",
                marker={'size': 10,
                        'color': 'green',
                        'opacity': 0.6,
                       });

iplot(fig)
```
